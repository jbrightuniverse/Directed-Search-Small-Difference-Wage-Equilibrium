```python
import matplotlib.pyplot as plt
import numpy as np
from io import BytesIO
from PIL import Image

def pi_h(w1, w2, gamma):
    #if (gamma*w1/2 + (1-gamma)*w1 > w2):
    #    return 1
    return ((gamma-2)*w2 + 2*w1)/(gamma*(w1+w2))
    #return 0

def inner_pi_l(w1, w2, gamma):
    if (round(gamma*w1/2 + (1-gamma)*w1, 8) > w2):
        return 1
        #return (2*(gamma-1)*w1 + (gamma + 1)*w2)/((gamma-1)*(w1+w2))
    #return ((3-gamma)*w2 - 2*w1)/((1-gamma)*(w1+w2))
    return 0

def pi_l(w1, w2, gamma):
    pi = inner_pi_l(w1, w2, gamma)
    if (pi > 1): 
        return 1
    elif (pi < 0): 
        return 0
    elif (gamma*w1/2 + (1-gamma)*w1 >= w2):
        return 3
    else: return 2
    #return pi
```

```python


images = []
for gam in np.linspace(0, 0.9999, 50):
    fig, ax = plt.subplots()
    frst = []
    fbb = []
    snd = []
    sbb = []
    for w in np.linspace(0.00000001, 1, 100):
        for w2 in np.linspace(0.00000001, 1, 100):
            ph = pi_h(w, w2, gam)
            if ph == 1:
                frst.append(w)
                fbb.append(w2)
            else:
                snd.append(w)
                sbb.append(w2)


    plt.scatter(frst, fbb, label = "Pi_H = 1")
    plt.scatter(snd, sbb, label = "Pi_H = not 1")

    plt.xlabel("w1")
    plt.ylabel("w2")

    ws = np.linspace(0.00000001, 1, 200)
    plt.plot(ws, ws, label = "w1 = w2", color = "green")
    plt.plot(ws, ws/2, label = "w1/2 = w2", color = "purple")
    plt.plot(ws, ws*(1-gam/2), label = "w2 = lambdathing", color = "red")
    fig.suptitle(f"w1 vs w2 in Pi_H for lambda = {round(gam, 2)}")
    ax.legend()
    filex = BytesIO()
    fig.savefig(filex, format = "png")
    plt.close()
    filex.seek(0)
    images.append(filex)
    

imgs = [Image.open(image) for image in images]
masterfilex = BytesIO()
imgs[0].save(fp=masterfilex, format='GIF', append_images=imgs[1:], save_all=True, duration=200, loop=0)

# this is a GIF-displaying hack from https://stackoverflow.com/a/32108899
from IPython.display import HTML
from base64 import b64encode
HTML("<img src='data:image/gif;base64,{0}'/>".format(b64encode(masterfilex.getvalue()).decode('utf-8')))
```
